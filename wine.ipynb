```python
import numpy as np
import pandas as pd

#quick check on datatypes within
winelist = pd.read_csv("Data/WineQT.csv")

#data is entirely numeric so unnecessary to convert this specific set
#find dataset with one for practice later
#Let's drop Id number since it is unnecessary
winelist = winelist.drop(columns = ['Id'])

#split into test/training sets
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(winelist, train_size = 0.7)
df_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df_train[df_train.columns] = scaler.fit_transform(df_train[df_train.columns])

df_train #all data except ID number are now relatively scaled
```

```python
#dividing training data set into X and Y (for quality)
y_train = df_train.pop('quality')
x_train = df_train

#build a linear model

import statsmodels.api as sm
x_train_lm = sm.add_constant(x_train)

lr_1 = sm.OLS(y_train, x_train_lm).fit()

lr_1.summary()
```

```python
#before removing high p-values, calculate Variance Inflation Factor for multicollinearity
# VIF = 1/(1-Ri^2)
from statsmodels.stats.outliers_influence import variance_inflation_factor

#create dataframe with all feature variables + VIFs
vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Let's drop density as it has a high VIF and p-value
X = x_train.drop(columns = ['density'])

#build fitted model again
x_train_lm = sm.add_constant(X)
lr_2 = sm.OLS(y_train, x_train_lm).fit()

print(lr_2.summary())

#calculate new vif
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
